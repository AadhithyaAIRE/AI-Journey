Now to know about the top 10% if gross amount holders.

```python
gross_projects=projects.sort_values('Gross Amount',ascending = False)
gross_projects=gross_projects.head(int(.005*len(gross_projects)))
print(len(gross_projects))
```

```python
# Group by 'Name' and calculate the total Gross Amount for each name in the top 10%
gross_share_by_name = gross_projects.groupby('Owner Name')['Gross Amount'].sum()

# Calculate the percentage share of each name
gross_share_by_name_percent = (gross_share_by_name / gross_share_by_name.sum()) * 100

# Sort the results to make the visualization more intuitive
gross_share_by_name_percent = gross_share_by_name_percent.sort_values(ascending=False)
```

```python
# Plot a pie chart of the share of Gross Amount by Name
plt.figure(figsize=(8, 8))
plt.pie(gross_share_by_name_percent, labels=gross_share_by_name_percent.index, autopct='%1.1f%%', startangle=140)
plt.title('Share of Gross Amount by Name (Top 10%)')
plt.show()
```
